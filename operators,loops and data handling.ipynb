```python
# Simple Assignment Operator (=)
x = 5
print("= Operator:", x)  # assigns 5 to x
```

```python
# Add AND Operator (+=)
x = 5
x += 3  # same as x = x + 3
print("+= Operator:", x)
```

```python
#Subtract AND Operator (-=)

x = 5
x -= 3  # same as x = x - 3
print("-= Operator:", x)
```

```python
# Multiply AND Operator (*=)
x = 5
x *= 3  # same as x = x * 3
print("*= Operator:", x)
```

```python
# Divide AND Operator (/=)
x = 9
x /= 3  # same as x = x / 3
print("/= Operator:", x)
```

```python
# Modulus AND Operator (%=)
x = 10
x %= 3  # same as x = x % 3
print("%= Operator:", x)
```

```python
# Floor Division AND Operator (//=)
x = 10
x //= 3  # same as x = x // 3
print("//= Operator:", x)
```

```python
# Exponent AND Operator (**=)
x = 2
x **= 3  # same as x = x ** 3
print("**= Operator:", x)
```

```python
# Bitwise AND Operator (&=)
x = 7   # binary: 111
x &= 3  # binary: 011 → result = 3
print("&= Operator:", x)
```

```python
# Bitwise OR Operator (|=)
x = 7   # binary: 111
x |= 3  # binary: 011 → result = 7
print("|= Operator:", x)
```

```python
# Bitwise XOR Operator (^=)
x = 7   # binary: 111
x ^= 3  # binary: 011 → result = 4
print("^= Operator:", x)
```

```python
# Bitwise Right Shift Operator (>>=)
x = 8   # binary: 1000
x >>= 2  # shifts bits right by 2 → binary: 0010 → result = 2
print(">>= Operator:", x)
```

```python
# Bitwise Left Shift Operator (<<=)
x = 3   # binary: 0011
x <<= 2  # shifts bits left by 2 → binary: 1100 → result = 12
print("<<= Operator:", x)
```

```python
# Equal To Operator (==)
x = 10
y = 10
print("== Operator:", x == y)  # True because both values are equal
```

```python
# Not Equal To Operator (!=)
x = 10
y = 5
print("!= Operator:", x != y)  # True because values are not equal
```

```python
# Greater Than Operator (>)
x = 15
y = 10
print("> Operator:", x > y)  # True because 15 is greater than 10
```

```python
# Less Than Operator (<)
x = 8
y = 12
print("< Operator:", x < y)  # True because 8 is less than 12
```

```python
# Greater Than or Equal To Operator (>=)
x = 10
y = 10
print(">= Operator:", x >= y)  # True because 10 is equal to 10
```

```python
# Less Than or Equal To Operator (<=)
x = 7
y = 10
print("<= Operator:", x <= y)  # True because 7 is less than 10
```

```python
# Logical AND Operator (and)
x = 10
y = 20
print("and Operator:", x > 5 and y > 15)  
# True because both conditions are True
```

```python
# Logical OR Operator (or)
x = 10
y = 5
print("or Operator:", x > 15 or y < 10)  
# True because one of the conditions is True
```

```python
# Logical NOT Operator (not)
x = 10
print("not Operator:", not(x > 5))  
# False because x > 5 is True, and not(True) becomes False
```

```python
# is Operator
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print("is Operator (x is z):", x is z)
# True because z and x refer to the same object

print("is Operator (x is y):", x is y)
# False because x and y are different objects (even though they have the same content)
```

```python
# is not Operator
print("is not Operator (x is not y):", x is not y)
# True because x and y are different objects
```
