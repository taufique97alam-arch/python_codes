```python
import numpy as np
a=np.array([(3,-3,4),(1,0,6),(-7,2,-5)])
b=np.array([(6,0,0),(-4,-1,8),(5,-1,8)])
#matrix
print("A=",a)
print("B=",b)
#addition
print("addition of A and B:-",a+b)
#multiplication
print("multiplication of A and B:-",np.dot(a,b))
#diagonal
print("diagonal of matrix A :-",np.diagonal(a))
print("diagonal of matrix b :-",np.diagonal(b))
#transpose
print("trandpose of matrix A :-",np.transpose(a))
print("transpose of matrix b :-",np.transpose(b))
#inverse
try:
  print("inverse of matrix A :-",np.linalg.inv(a))
except:
  print("matrix A is singular")
try:
  print("inverse of matrix b :-",np.linalg.inv(b))
except:
  print("matrix b is singular")
#rank
print("rank of matrix A :-",np.linalg.matrix_rank(a))
print("rank of matrix b :-",np.linalg.matrix_rank(b))
#determinant
print("determinant of matrix A :-",np.linalg.det(a))
print("determinant of matrix b :-",np.linalg.det(b))
```
