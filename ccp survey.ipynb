```python
import numpy as np
import pandas as pd
data={
    'employ': ['ram','shyam','rohit','abdul','aman',None,'ali','raju','mohit','akif'],
     'salary': [1000,2000,8000,1300,1700,np.nan,2200,1900,6000,1500],
      'id': [101,102,103,104,105,106,107,108,109,110]
}
df=pd.DataFrame(data)
print ("original dataset")
df
```

```python
df['employ']=df['employ'].fillna('unknown')
df['salary']=df['salary'].fillna(df['salary'].mean())
df['employ']=df['employ'].str.strip().str.title()

print('cleaned dastaset')
df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data={
    'employ_id': range(0,10,),
     'salary': [1000,2000,8000,1300,1700,2300,2200,1900,6000,1500]
}
df=pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['salary'], color='red')
plt.title("boxplot for salary(before outlier treatment)")
plt.show()
           
```

```python
q1=df['salary'].quantile(0.25)
q3=df['salary'].quantile(0.75)
qd=q3-q1
lower_limit=q1-1.5*qd
upper_limit=q3+1.5*qd
print(f"lower limit: {lower_limit}, upper limit: {upper_limit}")
outliers=df[(df['salary']<lower_limit)| (df['salary']>upper_limit)]
print("\ndetected outliers:")
outliers
```

```python
no_outliers=df[(df['salary']>=lower_limit)&(df['salary']<=upper_limit)]
print('dataset after removing outlier')
no_outliers
```
