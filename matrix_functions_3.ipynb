```python
#NAME : TAUFIQUE ALAM
#SYSTEM ID : 2025313333
#practical-5
import numpy as np
#define vectors
a=np.array([2,4,5])
b=np.array([5,7,3])
#1. check orthogonality
dot_ab=np.dot(a,b)
print("dot product:",dot_ab)
if dot_ab==0:
  print("vectors are orthogonal")
else:
  print("vectors are not orthogonal")
#2. normalize vectors
a_norm=a/np.linalg.norm(a)
b_norm=b/np.linalg.norm(b)
print("Normalized a:",a_norm)
print("Normalized b:",b_norm)
#3. check orthonormality
if np.isclose(np.dot(a_norm, b_norm),0) and np.isclose(np.linalg.norm(a_norm),1) and np.isclose(np.linalg.norm(b_norm),1):
  print("vectors are orthonormal")
else:
  print("vectors are not orthonormal")
```

```python
#NAME:TAUFIQUE ALAM
#SYSTEM ID:2025313333
#practical 8: principal component analysis(PCA)
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
#load datasets
data=load_iris()
x=data.data
#PCA
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x)
#plot
plt.scatter(x_pca[:,0],x_pca[:,1],c=data.target)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA if iris dataset')
plt.show()
```

```python
#NAME:TAUFIQUE ALAM
#SYSTEM ID:2025313333
#practical-10: constrained optimization with lagrange multipliers
import sympy as sp
#define variables
x,y,l=sp.symbols('x y lamda')
f= x**2 + y**2
g= x+y-1
#lagrange function
L= f-l*g
#solve system
sol=sp.solve([sp.diff(L,x),sp.diff(L,y),sp.diff(L,l)],(x,y,l))
print('optimal solution:',sol)
```
