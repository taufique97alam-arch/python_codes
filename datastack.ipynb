```python
l=[]
while True:
 a=int(input('''
            1.to push(enqueue)  
            2.to pop(dequeue)
            3.front queue
            4.rare queue
            5.display queue
            6.exit
            
            choose anyone:-'''))
 if a==1:
     b=input("enter the value:-")
     l.append(b)
     print(l)
 elif a==2:
     if len(l)==0:
      print("empty queue")
     else:
      del l[0]
      print(a)
 elif a==3:
     if len(l)==0:
      print("empty queue")
     else:
      print("front queue value",l[0])
 elif a==4:
     if len(l)==0:
      print("empty queue")
     else:
      print("rare queue value",l[-1])
 elif a==5:
     print("display queue value",l)
 else:
     break
```
