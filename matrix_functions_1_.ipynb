```python
#NAME:TAUFIQUE ALAM
#SYSTEM ID:2025313333
import numpy as np
a=np.array([(3,-3,4),(1,0,6),(-7,2,-5)])
b=np.array([(6,0,0),(-4,-1,8),(5,-1,8)])
#matrix
print("A=",a)
print("B=",b)
#addition
print("addition of A and B:-",(a+b))
#multiplication
print("multiplication of A and B:-",np.dot(a,b))
#diagonal
print("diagonal of matrix A :-",np.diagonal(a))
print("diagonal of matrix b :-",np.diagonal(b))
#transpose
print("trandpose of matrix A :-",np.transpose(a))
print("transpose of matrix b :-",np.transpose(b))
#inverse
try:
  print("inverse of matrix A :-",np.linalg.inv(a))
except:
  print("matrix A is singular")
try:
  print("inverse of matrix b :-",np.linalg.inv(b))
except:
  print("matrix b is singular")
#rank
print("rank of matrix A :-",np.linalg.matrix_rank(a))
print("rank of matrix b :-",np.linalg.matrix_rank(b))
#determinant
print("determinant of matrix A :-",np.linalg.det(a))
print("determinant of matrix b :-",np.linalg.det(b))
```

```python
#NAME:TAUFIQUE ALAM
#SYSTEM ID:2025313333
#vector operation and similarity
import numpy as np
#define vector
v1=np.array([2,4,6])
v2=np.array([1,3,5])
#1. dot product
dot_product=np.dot(v1,v2)
print("dot product=",dot_product)
#2. scalar multiplication
scalar_mul=2*v1
print("scalar multiplicatio=",scalar_mul)
#3. cosine similarity (using numPy)
cos_sim_numpy=np.dot(v1,v2)/ (np.linalg.norm(v1)*np.linalg.norm(v2))
print("cosine similarity (using numPy)=",cos_sim_numpy)
#cosine similarity (manual formula)
dot_manual=sum(v1[i]*v2[i]for i in range(len(v1)))
norm_v1=(sum(v1[i]**2 for i in range(len(v1))))**0.5
norm_v2=(sum(v2[i]**2 for i in range(len(v2))))**0.5
cos_sim_manual=dot_manual/(norm_v1 * norm_v2)
print("cosine similarity (manual formula)=",cos_sim_manual)
```
