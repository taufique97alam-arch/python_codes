```python
import numpy as np
a=np.array([1,-3,7])
b=np.array([-4,0,2])
c=np.dot(a,b)
if c==0:
  print("vectors are otrthogonal")
else:
  print("vectors are not orthogonal")
#2. normalize vectors
a_norm=a/np.linalg.norm(a)
b_norm=b/np.linalg.norm(b)
print("Normalized a:",a_norm)
print("Normalized b:",b_norm)
#3. check orthonormality
if np.isclose(np.dot(a_norm, b_norm),0) and np.isclose(np.linalg.norm(a_norm),1) and np.isclose(np.linalg.norm(b_norm),1):
  print("vectors are orthonormal")
else:
  print("vectors are not orthonormal")
```
